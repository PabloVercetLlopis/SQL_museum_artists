```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
```

# Cargar los csv

```python
df_museums = pd.read_csv("museum.csv")
df_paintings = pd.read_csv("work.csv", sep=';')
df_artists = pd.read_csv("artist.csv")
```

# Crear una base de datos SQLite en memoria

```python
engine = create_engine("sqlite://", echo=False)
```

# Subir los DataFrames como tablas SQL

```python
df_museums.to_sql('museums', engine, index=False, if_exists='replace')
df_paintings.to_sql('paintings', engine, index=False, if_exists='replace')
df_artists.to_sql('artists', engine, index=False, if_exists='replace')
```

# Numero de obras por estilo

```python
consulta_estilos = """
select distinct style, count(style) numero_estilos from paintings group by style order by numero_estilos DESC;
"""
df_estilos = pd.read_sql_query(consulta_estilos, engine)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="style", y="numero_estilos", data=df_estilos,  palette="Set2")  
plt.title("Número de obras por estilo")
plt.xlabel("Estilos")
plt.ylabel("Número de obras")
plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```

# Numero de artistas por Nacionalidad

```python
consulta_nacionalidades = """
select nationality, count (nationality)as numero_veces from artists group by nationality order by numero_veces desc;
"""
df_nacionalidades = pd.read_sql_query(consulta_nacionalidades, engine)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="nationality", y="numero_veces", data=df_nacionalidades,  palette="Set2")  
plt.title("Número de artistas por nacionalidad")
plt.xlabel("Nacionalidades")
plt.ylabel("Número de artistas")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

# Los 5 artistas con mas pinturas

```python
consulta_artistas_top = """
select full_name, count(paintings.name) as Numero_pinturas from artists join paintings on artists.artist_id=paintings.artist_id
group by full_name 
order by numero_pinturas DESC
limit 5;
"""
df_top_artistas = pd.read_sql_query(consulta_artistas_top, engine)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="full_name", y="Numero_pinturas", data=df_top_artistas, palette="viridis")
plt.title("Top 5 artistas con más pinturas")
plt.xlabel("Artista")
plt.ylabel("Número de pinturas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Estilo de pintura más común por nacionalidad

```python
consulta_estilo_nacionalidad = """
SELECT nationality, style, numero_veces_estilo
FROM (
  SELECT 
    artists.nationality,
    paintings.style,
    COUNT(paintings.style) AS numero_veces_estilo,
    ROW_NUMBER() OVER (PARTITION BY artists.nationality ORDER BY COUNT(paintings.style) DESC) AS fila
  FROM 
    artists
  JOIN 
    paintings ON artists.artist_id = paintings.artist_id
  GROUP BY 
    artists.nationality, paintings.style
) AS subconsulta 
WHERE 
  fila = 1;
"""
df_estilo_nacionalidad = pd.read_sql_query(consulta_estilo_nacionalidad, engine)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(
    x="numero_veces_estilo", 
    y="nationality", 
    hue="style", 
    data=df_estilo_nacionalidad, 
    dodge=False, 
    palette="Set2"
)
plt.title("Estilo de pintura más frecuente por nacionalidad")
plt.xlabel("Número de veces")
plt.ylabel("Nacionalidad")
plt.legend(title="Estilo")
plt.tight_layout()
plt.show()
```

# Museos con la mayor diversidad de estilos

```python
consulta_diversidad_estilos = """
select m.name, count(style) as estilos 
from museums m join paintings p  on m.museum_id = p.museum_id
group by m.museum_id, m.name
order by estilos DESC 
limit 10;
"""
df_diversidad_estilos = pd.read_sql_query(consulta_diversidad_estilos, engine)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='estilos', y='name', data=df_diversidad_estilos, palette = 'Set2')
plt.xlabel('Número de estilos')
plt.ylabel('Museos')
plt.title('Museos con la mayor diversidad de estilos artísticos')
plt.tight_layout()
plt.show()
```

# Ciudades con mas de un museo 

```python
consulta_ciudades = """
select city, count(name) as numero_museos from museums m 
group by city
having count(name) > 1
order by numero_museos desc
"""
df_ciudades = pd.read_sql_query(consulta_ciudades, engine)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='numero_museos', y='city', data = df_ciudades, palette = 'Set2')
plt.xlabel('Número de museos')
plt.ylabel('Ciudades')
plt.title('Ciudades con más de 1 museo')
plt.xticks(np.arange(1, 5, 1)) 
plt.tight_layout()
plt.show()
```

# Rangos de nacimientos de los artista por nacionalidades

```python
query = """
SELECT 
  nationality,
  MIN(birth) AS año_más_antiguo,
  MAX(birth) AS año_más_reciente
FROM artists
GROUP BY nationality;
"""
df_rango_nacimientos = pd.read_sql_query(query, engine)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
for i, row in df_rango_nacimientos.iterrows():
    plt.plot(
        [row['año_más_antiguo'], row['año_más_reciente']],
        [row['nationality']] * 2,
        marker='o',
        color='steelblue'
    )

plt.title("Rango de nacimiento de artistas por nacionalidad")
plt.xlabel("Año de nacimiento")
plt.ylabel("Nacionalidad")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

# Estilos que se encuentran en mas de 1 museo

```python
consulta_estilos_museos = """
select style, count(style) as esta_en_museos 
from paintings join museums on paintings.museum_id = museums.museum_id
group by style
having count (DISTINCT paintings.museum_id) > 1
"""
df_estilos_museos = pd.read_sql_query(consulta_estilos_museos, engine)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data = df_estilos_museos, x='esta_en_museos', y='style', palette='YlGnBu')
plt.title('Número de Museos en los que se encuentran los Estilos de Pintura')
plt.ylabel('Estilo')
plt.xlabel('Número de Museos')
plt.tight_layout()
plt.grid(True)
plt.show()
```
